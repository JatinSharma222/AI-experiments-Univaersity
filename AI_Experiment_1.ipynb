# ***Performing data cleaning i.e handling missing values, outliers, and duplicates***

```python
# Check for missing values in each column to identify potential areas for data cleaning
df.isnull().sum()
```

```python
df["Item_Weight"].mean()
```

```python
df["Item_Weight"].fillna(df["Item_Weight"].mean(),inplace=True)
```

```python
# Check for missing values in each column to identify potential areas for data cleaning
df.isnull().sum()
```

```python
df["Outlet_Size"].mode()
```

```python
# Display the first five rows of the dataset to understand its structure and preview the data
df.head()
```

```python
mode_outlet_size = df.pivot_table(values="Outlet_Size", columns = "Outlet_Type", aggfunc=(lambda x: x.mode()[0]))
```

```python
print(mode_outlet_size)
```

```python
miss_value = df["Outlet_Size"].isnull()
```

```python
print(miss_value)
```

```python
df.loc[miss_value, "Outlet_Size"] = df.loc[miss_value, "Outlet_Type"].apply(lambda x: mode_outlet_size[x])
```
